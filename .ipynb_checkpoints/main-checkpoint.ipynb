### Randomly Sampled Texture

```python
def quilt_random(sample, out_size, patch_size):
    """
    Randomly samples square patches of size patchsize 
    from sample in order to create an output image of 
    size outsize.
    """
    assert sample.shape[1] - patch_size - 1 > 0 and sample.shape[0] - patch_size - 1 > 0
    out = np.zeros((out_size))
    for x in np.arange(0, out_size[1], patch_size):
        for y in np.arange(0, out_size[0], patch_size):
            r_y = np.random.randint(0, sample.shape[1] - patch_size - 1)
            r_x = np.random.randint(0, sample.shape[0] - patch_size - 1)
            if y + patch_size > out_size[0] or x + patch_size > out_size[1]:
                continue
            patch = sample[r_x:r_x+patch_size, r_y:r_y+patch_size]
            out[y:y+patch_size, x:x+patch_size] = patch/255
    return out
```

```python
out_size = [400, 400, 3]
patch_size = 21
quilted = quilt_random(texture, out_size, patch_size)
```

### Overlapping Patches

```python
@jit
def ssd(B_i, S_B, overlap, orientation, j=0):
    ssd = []
    if orientation == 0:
        for B_j in S_B:
            ssd.append(np.sum((B_j[:,0:overlap] - B_i[:,0:overlap]) ** 2))
    if orientation == 1:
        for B_j in S_B:
            ssd.append(np.sum((B_j[0:overlap, :] - B_i[0:overlap, :]) ** 2))
    if orientation == 2:
        for B_j in S_B:
            ssd.append(np.sum((B_j[0:overlap, :] - B_i[0:overlap, :])) ** 2 + np.sum((B_j[overlap:,0:overlap] - B_i[overlap:,0:overlap]) ** 2))
    return np.asarray(ssd)

def choose_sample(ssds, j=0):
    minc = np.amin(ssds)
    if minc == 0:
        minc = max(minc, 1e-4)
    idxs = np.where(ssds < minc * (1 + tol))[0]
    idx = np.random.choice(idxs)
    return idx

def mask(patch_size, overlap, orientation):
    whole = np.zeros([patch_size, patch_size, 3])
    if orientation == 0:
        # only left (first row)
        whole[0:whole.shape[0], 0:overlap] = [1, 1, 1] 
        return whole
    elif orientation == 1:
        # only top (first column)
        whole[0:overlap, 0:whole.shape[1]] = [1, 1, 1]
        return whole
    # the usual case
    whole[0:overlap, 0:whole.shape[1]] = [1, 1, 1]
    whole[overlap:whole.shape[0], 0:overlap] = [1, 1, 1]
    return whole

@jit
def fill_first_row(patch_size, overlap, S_B, out, tol):
    for i in np.arange(patch_size - overlap, out.shape[1] - patch_size, patch_size - overlap):
        B_i = out[0:patch_size, i:i+patch_size]
        ssds = ssd(B_i, S_B, overlap, 0)
        idx = choose_sample(ssds)
        B_j = S_B[idx]
        out[0:patch_size, i:i+patch_size] = B_j 

@jit
def fill_first_col(patch_size, overlap, S_B, out, tol):
    for i in np.arange(patch_size - overlap, out.shape[0] - patch_size, patch_size - overlap):
        B_i = out[i:i+patch_size, 0:patch_size]
        ssds = ssd(B_i, S_B, overlap, 1)
        idx = choose_sample(ssds)
        B_j = S_B[idx]
        out[i:i+patch_size, 0:patch_size] = B_j

@jit
def fill(patch_size, overlap, S_B, out, tol):
    for x in np.arange(patch_size - overlap, out.shape[1] - patch_size, patch_size - overlap):
        for y in np.arange(patch_size - overlap, out.shape[0] - patch_size, patch_size - overlap):
            B_i = out[x:x+patch_size, y:y+patch_size]
            ssds = ssd(B_i, S_B, overlap, 2)
            idx = choose_sample(ssds)
            B_j = S_B[idx]
            out[x:x+patch_size, y:y+patch_size] = B_j

def quilt_simple(sample, out_size, patch_size, overlap, tol):
    out = np.zeros(out_size)
    # Compute all patches
    S_B = image.extract_patches_2d(sample, (patch_size, patch_size))
    # Place first patch in top left
    r = np.random.randint(0, S_B.shape[0])
    patch = S_B[r]
    out[0:patch_size, 0:patch_size] = patch
    # First row
    fill_first_row(patch_size, overlap, S_B, out, tol)
    # First col
    fill_first_col(patch_size, overlap, S_B, out, tol)
    # Fill rest
    fill(patch_size, overlap, S_B, out, tol)
    return out
```

### Seam Finding

```python
def get_energy(img):
    """
    Calculate energy of image.
    """
    del_x = np.square(np.roll(img, -1, axis = 0) - np.roll(img, 1, axis = 0))
    del_y = np.square(np.roll(img, -1, axis = 1) - np.roll(img, 1, axis = 1))
    energy = np.sum(del_x, axis = 2) + np.sum(del_y, axis = 2)
    return energy

@jit
def get_m_mat(energy, w, h):
    """
    Single calculation of the DP matrix for the approximation algorithm.
    """
    m_mat = energy.copy()
    for r in np.arange(1, h):
        for c in np.arange(0, w):
            if c == 0:
                m_mat[r][c] = energy[r][c] + min(m_mat[r-1][c], m_mat[r-1][c+1])
            elif c == w - 1:
                m_mat[r][c] = energy[r][c] + min(m_mat[r-1][c-1], m_mat[r-1][c])
            else:
                m_mat[r][c] = energy[r][c] + min(m_mat[r-1][c-1], m_mat[r-1][c], m_mat[r-1][c+1])
    return m_mat

@jit
def find_vertical_seam(m_mat):
    """
    An approximation of the Optimal seam carving solution, by not recalulating
    the DP matrix at each carve. Performs recognizably worse than the Optimal
    solution. About 3x speed up.
    """
    seam = np.array([])
    min_idx = np.nonzero(m_mat[len(m_mat) - 1] == min(m_mat[len(m_mat) - 1]))[0][0]
    seam = np.append(seam, min_idx)
    n = len(m_mat)
    l = min_idx if min_idx == 0 else min_idx - 1
    r = min_idx if min_idx == len(m_mat[0]) - 1 else min_idx + 1
    m = min_idx
    for row in np.arange(n - 1, 0, -1):
        min_val = min([m_mat[row][l], m_mat[row][m], m_mat[row][r]])
        if min_val == m_mat[row][l]:
            seam = np.append(seam, l)
            l = l if l == 0 else l - 1
            r = l if l == len(m_mat[0]) - 1 else l + 1
            m = l
        elif min_val == m_mat[row][r]:
            seam = np.append(seam, r)
            l = r if r == 0 else r - 1
            r = r if r == len(m_mat[0]) - 1 else r + 1
            m = r
        else:
            seam = np.append(seam, m)
            l = m if m == 0 else m - 1
            r = m if m == len(m_mat[0]) - 1 else m + 1
            m = m
    return seam
    
def find_horizontal_seam(img, energy, w, h, m_mat):
    img2 = np.rot90(img, 3)
    energy2 = np.rot90(energy, 3)
    h2, w2 = w, h
    m_mat = get_m_mat(energy2, w2, h2)
    seam = find_vertical_seam(m_mat)
    return seam

@jit
def mask_top(img1, img2):
    a = img1[0:overlap, :]
    b = img2[0:overlap, :]
    e = get_energy(a)
    w = a.shape[1]
    h = a.shape[0]
    m_mat = get_m_mat(e, w, h)
    seam = find_horizontal_seam(a, e, w, h, m_mat).astype(int)
    for r in range(img1.shape[0]):
        img1[:, r][:seam[r]] = img2[:, r][:seam[r]]

@jit
def mask_left(img1, img2, j=0):
    a = img1[:, 0:overlap]
    b = img2[:, 0:overlap]
    e = get_energy(a)
    w = a.shape[1]
    h = a.shape[0]
    m_mat = get_m_mat(e, w, h)
    seam = find_vertical_seam(m_mat).astype(int)
    for r in range(img1.shape[1]):
        img1[r][:seam[r]] = img2[r][:seam[r]]
```

```python
@jit
def fill_first_row_cut(patch_size, overlap, S_B, out, tol):
    j = 1
    for i in np.arange(patch_size - overlap, out.shape[1] - patch_size, patch_size - overlap):
        B_i = out[0:patch_size, i:i+patch_size]
        ssds = ssd(B_i, S_B, overlap, 0)
        idx = choose_sample(ssds, j)
        B_j = S_B[idx]
        mask_left(B_j, B_i, j)
        out[0:patch_size, i:i+patch_size] = B_j 
        j -= 1

@jit
def fill_first_col_cut(patch_size, overlap, S_B, out, tol):
    for i in np.arange(patch_size - overlap, out.shape[0] - patch_size, patch_size - overlap):
        B_i = out[i:i+patch_size, 0:patch_size]
        ssds = ssd(B_i, S_B, overlap, 1)
        idx = choose_sample(ssds)
        B_j = S_B[idx]
        mask_top(B_j, B_i)
        out[i:i+patch_size, 0:patch_size] = B_j

@jit
def fill_cut(patch_size, overlap, S_B, out, tol):
    for x in np.arange(patch_size - overlap, out.shape[1] - patch_size, patch_size - overlap):
        for y in np.arange(patch_size - overlap, out.shape[0] - patch_size, patch_size - overlap):
            B_i = out[x:x+patch_size, y:y+patch_size]
            ssds = ssd(B_i, S_B, overlap, 2)
            idx = choose_sample(ssds)
            B_j = S_B[idx]
            mask_left(B_j, B_i)
            mask_top(B_j, B_i)
            out[x:x+patch_size, y:y+patch_size] = B_j

def quilt_cut(sample, out_size, patch_size, overlap, tol):
    out = np.zeros(out_size)
    # Compute all patches
    S_B = image.extract_patches_2d(sample, (patch_size, patch_size))
    # Place first patch in top left
    r = np.random.randint(0, S_B.shape[0])
    patch = S_B[r]
    out[0:patch_size, 0:patch_size] = patch
    # First row
    fill_first_row_cut(patch_size, overlap, S_B, out, tol)
    # First col
    fill_first_col_cut(patch_size, overlap, S_B, out, tol)
    # Fill rest
    fill_cut(patch_size, overlap, S_B, out, tol)
    return out
```

```python
out_size = [200, 200, 3]
patch_size = 31
overlap = int(patch_size / 4)
tol = .1
start = time.time()
a = quilt_cut(texture, out_size, patch_size, overlap, tol)
end = time.time()
print("Elapsed (with compilation) = %s" % (end - start))
```

### Texture Transfer

```python
# @jit
# def calc_ssd_tex(B_i, S_B, S_C, i, T_i, a, orientation, it, out, x, y):
#     prev_ssd = 0
#     patch_size = B_i.shape[0]
#     B_j = S_B[i]
#     B_c = S_C[i]
#     if it > 0:
#         prev_ssd = np.sum((S_B - out[x:x+patch_size, y:y+patch_size]) ** 2, axis=(1, 2, 3))
#     ssd = 0
#     style_ssd = np.sum((S_C - T_i) ** 2, axis=(1, 2, 3))
#     if orientation == 0:
#         overlap_ssd = np.sum((S_B[:,:,0:overlap] - B_i[:,0:overlap]) ** 2, axis=(1, 2, 3))
#         ssd = a * (overlap_ssd + prev_ssd) + (1 - a) * style_ssd
#         return ssd
#     elif orientation == 1:
#         overlap_ssd = np.sum((S_B[:,0:overlap, :] - B_i[0:overlap, :]) ** 2, axis=(1, 2, 3))
#         ssd = a * (overlap_ssd + prev_ssd) + (1 - a) * style_ssd
#         return ssd
#     overlap_ssd = np.sum((S_B[:,0:overlap, :] - B_i[0:overlap, :]) ** 2, axis=(1, 2, 3))
#     overlap_ssd += np.sum((S_B[:,overlap:,0:overlap] - B_i[overlap:,0:overlap]) ** 2, axis=(1, 2, 3))
#     ssd = a * (overlap_ssd + prev_ssd) + (1 - a) * style_ssd
#     return ssd

# @jit
# def ssd_tex(B_i, S_B, S_C, T_i, overlap, orientation, a, it, out, x, y):
#     ssd = np.array([])
#     ssd = calc_ssd_tex(B_i, S_B, S_C, i, T_i, a, orientation, it, out, x, y)
#     return ssd

@jit
def calc_ssd_tex(B_i, S_B, S_C, i, T_i, a, orientation, it, out, x, y):
    prev_ssd = 0
    patch_size = B_i.shape[0]
    B_j = S_B[i]
    B_c = S_B[i]
    if it > 0:
        prev_ssd = np.sum((B_j - out[x:x+patch_size, y:y+patch_size]) ** 2)
    ssd = 0
    style_ssd = np.sum((B_c - T_i) ** 2)
    if orientation == 0:
        overlap_ssd = np.sum((B_j[:,0:overlap] - B_i[:,0:overlap]) ** 2)
        ssd = a * (overlap_ssd + prev_ssd) + (1 - a) * style_ssd
        return ssd
    elif orientation == 1:
        overlap_ssd = np.sum((B_j[0:overlap,:] - B_i[0:overlap,:]) ** 2)
        ssd = a * (overlap_ssd + prev_ssd) + (1 - a) * style_ssd
        return ssd
    overlap_ssd = np.sum((B_j[0:overlap,:] - B_i[0:overlap,:]) ** 2)
    overlap_ssd += np.sum((B_j[overlap:,0:overlap] - B_i[overlap:,0:overlap]) ** 2)
    ssd = a * (overlap_ssd + prev_ssd) + (1 - a) * style_ssd
    return ssd

@jit
def ssd_tex(B_i, S_B, S_C, T_i, overlap, orientation, a, it, out, x, y):
    ssd = []
    for i in range(len(S_B)):
        ssd.append(calc_ssd_tex(B_i, S_B, S_C, i, T_i, a, orientation, it, out, x, y))
    return np.asarray(ssd)
```

```python
@jit
def fill_first_row_tex(patch_size, overlap, S_B, texture, out, tol, a, it, S_C):
    for i in np.arange(patch_size - overlap, out.shape[1] - patch_size, patch_size - overlap):
        B_i = out[0:patch_size, i:i+patch_size]
        T_i = texture[0:patch_size, i:i+patch_size]
        ssds = ssd_tex(B_i, S_B, S_C, T_i, overlap, 0, a, it, out, 0, i)
        idx = choose_sample(ssds)
        B_j = S_B[idx]
        mask_left(B_j, B_i)
        out[0:patch_size, i:i+patch_size] = B_j 

@jit
def fill_first_col_tex(patch_size, overlap, S_B, texture, out, tol, a, it, S_C):
    for i in np.arange(patch_size - overlap, out.shape[0] - patch_size, patch_size - overlap):
        B_i = out[i:i+patch_size, 0:patch_size]
        T_i = texture[i:i+patch_size, 0:patch_size]
        ssds = ssd_tex(B_i, S_B, S_C, T_i, overlap, 1, a, it, out, i, 0)
        idx = choose_sample(ssds)
        B_j = S_B[idx]
        mask_top(B_j, B_i)
        out[i:i+patch_size, 0:patch_size] = B_j

@jit
def fill_tex(patch_size, overlap, S_B, texture, out, tol, a, it, S_C):
    for x in np.arange(patch_size - overlap, out.shape[0] - patch_size, patch_size - overlap):
        for y in np.arange(patch_size - overlap, out.shape[1] - patch_size, patch_size - overlap):
            B_i = out[x:x+patch_size, y:y+patch_size]
            T_i = texture[x:x+patch_size, y:y+patch_size]
            ssds = ssd_tex(B_i, S_B, S_C, T_i, overlap, 2, a, it, out, x, y)
            idx = choose_sample(ssds)
            B_j = S_B[idx]
            mask_left(B_j, B_i)
            mask_top(B_j, B_i)
            out[x:x+patch_size, y:y+patch_size] = B_j

def quilt_tex(sample, texture, out_size, patch_size, overlap, tol, a, corresp, it=0, prev=""):
    out = 0
    if it == 0:
        out = np.zeros(out_size)
    else:
        out = prev
    S_C = image.extract_patches_2d(corresp, (patch_size, patch_size))
    # Compute all patches
    S_B = image.extract_patches_2d(sample, (patch_size, patch_size))
    # Place first patch in top left
    r = np.random.randint(0, S_B.shape[0])
    patch = S_B[r]
    out[0:patch_size, 0:patch_size] = patch
    # First row
    fill_first_row_tex(patch_size, overlap, S_B, texture, out, tol, a, it, S_C)
    # First col
    fill_first_col_tex(patch_size, overlap, S_B, texture, out, tol, a, it, S_C)
    # Fill rest
    fill_tex(patch_size, overlap, S_B, texture, out, tol, a, it, S_C)
    return out
```

```python
texture = skio.imread("samples/bricks_small.jpg")
tex = (skio.imread("samples/al.jpg"))
corresp = (texture)
patch_size = int(np.amax(texture.shape) / 3.4)
if patch_size % 2 == 0:
    patch_size += 1
N = 3
tol = .1
a = .8 * ((0) / (N - 1)) + .1
overlap = int(patch_size / 6)
start = time.time()
b = quilt_tex(texture, tex, [tex.shape[0], tex.shape[1], 3], patch_size, overlap, tol, a, corresp, 0)
end = time.time()
print("Elapsed (with compilation) = %s" % (end - start))
skio.imshow(b/255)
skio.show()
skio.imsave("output/abe_iter_1.jpg", b)
for i in range(1, N):
    patch_size = int(patch_size / 2.5)
    if patch_size % 2 == 0:
        patch_size += 1
    overlap = int(patch_size / 6)
    a = .8 * ((i) / (N - 1)) + .1
    start = time.time()
    b = quilt_tex(texture, tex, tex.shape, patch_size, overlap, tol, a, corresp, i, b)
    end = time.time()
    print("Elapsed (with compilation) = %s" % (end - start))
    skio.imshow(b/255)
    skio.show()
    skio.imsave("output/abe_iter_" + str(i+1) + ".jpg", b)
```
